## Kapitola 1.4 Proměnné

Častokrát je velmi neefektivní používat hodnoty napřímo a lepším efektivnějším způsobem je využívat proměnných.<br>

Proměnnou je nejlepší představit si jako šuplík nebo krabičku, do které vložíte hodnotu a tento šuplík/krabičku si pojmenujete.
Poté přistupujete k hodnotě v proměnné díky tomuto vytvořenému názvu.<br>
Proměnnou vytvoříme následovně:
```název_proměnné = hodnota```

````python
leden_výplata = 10000
únor_výplata = 12000
celkem = leden_výplata + únor_výplata
celkem
````

Proměnné se nemohou jmenovat jakkoliv:
  * Nesmí začínat číslicí
  * Nesmí obsahovat speciální znaky
  * Při názvech složených z více slov je mezera k oddělení slov chyba a zpravidla se používá podtržítko tzv.snake_case

## Kapitola 1.5: Změna hodnoty proměnné

V některých případech je třeba měnit hodnoty proměnných v průběhu programu. python nám umožňuje takový zápis zkrátit.
Níže vidíte přehled zkráceného zápisu aritmetických operací i jejich klasickými variantami:

* A += výraz      ekvivalent výrazu A = A + výraz
* A -= výraz      ekvivalent výrazu A = A - výraz
* A *= výraz      ekvivalent výrazu A = A * výraz
* A /= výraz      ekvivalent výrazu A = A / výraz
* A //= výraz     ekvivalent výrazu A = A // výraz
* A %= výraz      ekvivalent výrazu A = A % výraz
* A **= výraz     ekvivalent výrazu A = A ** výraz

````python
A = 10
A += 20
A
````

````python
A = 10
A **= 2
A
````
