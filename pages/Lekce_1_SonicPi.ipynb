### Cvičení 1.4.1: Priority check

Rozhodni v buňce níže, zda se varianta A liší od varianty B.
Pokud ano, v čem? a proč?
Buňku si přehraj

````python
ton = 63

# Verze A
ton_A = ton + 12 / 2
play(ton_A)
sleep(1)

# Verze B
ton_B = (ton + 12) / 2
play(ton_B)
````

Né vždy potřebujeme nebo můžeme dělit tímto způsobem. Někdy je potřeba, aby výsledkem bylo zaručeně celé číslo. Toho můžeme docílit celočíselným dělením: _x // y_
Spusť buňku níže a poslechni si, zda slyšíš nějaký rozdíl.

````python
ton = 200
delitel = 3
vysledek_1 = ton/delitel # 200/3 = 66.6666...
vysledek_2 = ton//delitel # 200//3 = 66
play(vysledek_1)
sleep(1)
play(vysledek_2)
````

Proměnná _vysledek2_ ze cvičení výše má po celočíselném dělení hodnotu 66 celých zatímco proměnná _vysledek1 = 66.6666_

## Kapitola 1.5: Změna hodnoty proměnné

V některých případech je třeba měnit hodnoty proměnných v průběhu programu. python nám umožňuje takový zápis zkrátit.
Níže vidíte přehled zkráceného zápisu aritmetických operací i jejich klasickými variantami:

* A += výraz      ekvivalent výrazu A = A + výraz
* A -= výraz      ekvivalent výrazu A = A - výraz
* A *= výraz      ekvivalent výrazu A = A * výraz
* A /= výraz      ekvivalent výrazu A = A / výraz
* A //= výraz     ekvivalent výrazu A = A // výraz
* A %= výraz      ekvivalent výrazu A = A % výraz
* A **= výraz     ekvivalent výrazu A = A ** výraz

````python
ton = 10
ton += 40
play(ton)
````

Předchozí buňka se vykoná a na řádku 3 kde voláme funkci _play(ton)_ má proměnná _ton_ hodnotu 50
Zkusme si to ověřit vuz buňka níže

````python
ton = 10
play(ton)
sleep(1)
ton += 40
play(ton)
````

První play(ton) na řádku 2 se spustí, ale tím že je tam hodnota 10, naše ucho nejspíš nebude schopné nic slyšet, jelikož je tón pod hranicí slyšitelnosti, poté je vteřina pauza sleep(1), přičteme hodnotu 40 k původní hodnotě tónu a uložíme do proměnné ton a zavolá se na druhý play(ton) na řádku 5, který už slyšíme.
